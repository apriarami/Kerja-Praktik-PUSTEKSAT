```python
ra_sun = df3['RightAscension (deg)'].to_numpy(dtype='float64')
dec_sun = df3['Declination (deg)'].to_numpy(dtype='float64')
```

```python
df4 = pd.read_excel('drive/MyDrive/Kerja Praktik/LAPAN A3 orbital elements_RA DE Sun.xlsx',sheet_name='Orbital Elements')
df4.dtypes
```

```python
inc = df4['Inclination (deg)'].to_numpy(dtype='float64')
raan = df4['RAAN (deg)'].to_numpy(dtype='float64')
```

```python
ra_sun_rad = (np.pi/180)*ra_sun
dec_sun_rad = (np.pi/180)*dec_sun
inc_rad = (np.pi/180)*inc
raan_rad = (np.pi/180)*raan

inc_rad
```

```python
beta1 = np.zeros([len(ra_sun)])
beta2 = np.zeros([len(ra_sun)])
beta3 = np.zeros([len(ra_sun)])
beta4 = np.zeros([len(ra_sun)])
beta = np.zeros([len(ra_sun)])
for i in range(len(ra_sun)):
  beta1[i] = (np.cos(ra_sun_rad[i]))*(np.sin(raan_rad[i]))*(np.sin(inc_rad[i]))
  beta2[i] = (np.sin(ra_sun_rad[i]))*(np.cos(dec_sun_rad[i]))*(np.cos(raan_rad[i]))*(np.sin(inc_rad[i]))
  beta3[i] = (np.sin(ra_sun_rad[i]))*(np.sin(dec_sun_rad[i]))*(np.cos(inc_rad[i]))
  beta4[i] = beta1[i] - beta2[i] + beta3[i]
beta5 = np.arcsin(beta4)
beta = (180/np.pi)*beta5
```
